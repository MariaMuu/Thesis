```python
import pandas as pd
```

```python
train = pd.read_csv('/content/validation_results_20250623_105855.csv')
```

```python
test = pd.read_csv('/content/validation_results_20250623_123305.csv')
```

```python
train.head(10)
```

```python
train.columns
```

```python
train_col = train[['question', 'sparql', 'result_count']]
train_col
```

```python
test_col = test[['question', 'sparql', 'result_count']]
test_col
```

```python
train_filtered = train_col[train_col['result_count'] > 0]
train_filtered
```

```python
test_filtered = test_col[test_col['result_count'] > 0]
test_filtered
```

```python
# prompt: i want to reset the index of the dataset and get rid off the column result_count

train_reset = train_filtered.reset_index(drop=True)
train_cleaned = train_reset.drop(columns=['result_count'])
train_cleaned
```

```python
test_reset = test_filtered.reset_index(drop=True)
test_cleaned = test_reset.drop(columns=['result_count'])
test_cleaned
```

```python
train_cleaned.to_csv('train_cleaned.csv', index=False)
```

```python
test_cleaned.to_csv('test_cleaned.csv', index=False)
```
