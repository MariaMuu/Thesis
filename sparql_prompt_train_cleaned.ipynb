cleaning the data and keeping only two columns

```python
new_df = df[['question','sparql_wikidata']] # keeping only the necessary columns
new_df
```

We will add some words and symbols to our data in order to make them fit for the fine-tuning process

Preparing the column "question"

```python
prefix = '{"role":"user", "content": "' #first part
```

```python
suffix_symbols = '"},' #last part
```

```python
for col in new_df.columns:
  new_df['question'] = new_df['question'].astype(str).apply(lambda x: prefix + x.strip() if not x.startswith(prefix) else x + suffix_symbols)
```

Preparing the column "sparql_wikidata"

```python
prefix_new = '{"role": "assistant", "content": "'
```

```python
suffix_symbols_new = '"}]}'
```

```python
new_df['sparql_wikidata'] = new_df['sparql_wikidata'].astype(str).apply(lambda x: prefix_new + x.strip() if not x.startswith(prefix_new) else x + suffix_symbols_new)
```

```python
new_df
```

We will now add one more column that will contain the first part of each prompt

```python
input_words = '{"messages": [{"role": "system","content": "You are a SPARQL generator for Wikidata. Given a natural language question, return a syntactically correct and executable SPARQL query using the Wikidata ontology. Respond only with the query."},'
```

```python
prompt = [f"{input_words}" for i in new_df.index]
```

```python
new_df['prompt'] = prompt
```

```python
new_df
```

Putting them in the correct order

```python
new_df = new_df[['prompt', 'question', 'sparql_wikidata']]
```

```python
new_df
```

Merging the columns into one column

```python
new_df['final_prompt'] = new_df['prompt'].astype(str) + new_df['question'].astype(str) + new_df['sparql_wikidata'].astype(str)
```

```python
df_final = new_df[['final_prompt']]
df_final
```

Saving the new files

```python
df_final.to_csv('final_prompt_train.csv', index=False)
```

```python
df_final.to_json('final_prompt_train.json', orient='records', lines=True)
```
