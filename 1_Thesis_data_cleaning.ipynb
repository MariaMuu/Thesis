# **Preparing the data for the fine-tuning process**

**Importing python libraries**

```python
import pandas as pd
```

**Reading the files**

```python
train = pd.read_csv('/content/train.csv')
test = pd.read_csv('/content/test.csv')
```

```python
pd.set_option('display.max_colwidth', 1000) # in order to be able to see the content of the table
```

**Checking how many null values the are in our datasets**

```python
train.isna().sum()
```

**Drop the rows with the null values**

```python
train.dropna()
```

```python
test.isna().sum() # Doing the same for the validation set
```

```python
test.dropna()
```

```python
new_train = train[['question','sparql_wikidata']] # keeping only the necessary columns
new_train
```

```python
new_test = test[['question','sparql_wikidata']] # keeping only the necessary columns
new_test
```

**Keeping only 6000 rows for the training set and 1500 for the validation set**

```python
new_train_random = new_train.sample(n = 6000, random_state=42)
new_test_random = new_test.sample(n=1500, random_state=25)

new_train_random.info()
new_test_random.info()
```

```python
# checking the datasets

new_train_random.head()
new_test_random.head()
```

```python
# Reset the index of the dataframes
new_train_random = new_train_random.reset_index(drop=True)
new_test_random = new_test_random.reset_index(drop=True)

new_train_random.head()
new_test_random.head()
```

Saving Processed Data to CSV & Json files

```python
new_train_random.to_csv('thesis_prompt_train_random.csv', index=False)
new_test_random.to_csv('thesis_prompt_test_random.csv', index=False)
```

```python
new_train_random.to_json('thesis_prompt_train_random.json', orient='records', lines=True)
new_test_random.to_json('thesis_prompt_test.json_random', orient='records', lines=True)
```
